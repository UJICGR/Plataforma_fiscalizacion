```python
# -*- coding: utf-8 -*-
from urllib.request import urlretrieve
import textract
urlretrieve("https://cgrfiles.cgr.go.cr/publico/docs_cgr/2019/SIGYD_D_2019023031.pdf", "download.pdf")
text = textract.process("download.pdf")
text2=text.decode("utf-8")
text2
```

```python

import re
```

# Preparación del texto, elimina basura o palabras sin sentido para el análisis.

```python
#Limpieza de metadata
output_clean = text2.replace('\n', ' ')
output_clean = output_clean.replace('\r', ' ')
output_clean = output_clean.replace('....', ' ')
output_clean = output_clean.replace('Apdo', ' ')
output_clean = output_clean.replace('Licda', ' ')
output_clean = output_clean.replace('?', ' ')
output_clean = output_clean.replace('Asimismo', ' ')
```

```python

output_clean
```

```python
output_clean =re.sub(' +', ' ', output_clean)
```

```python
output_clean
```

# Importa el modelo pre-entrenado en español (https://spacy.io/models/es)

```python
import es_core_news_md
nlp = es_core_news_md.load()
```

# Aplicamos el  modelo a nuestro texto.

```python
doc = nlp(output_clean)
```

# Tratamos de obtener los sustativos.

```python
print("Sintagmas Nominales:", [chunk.text for chunk in doc.noun_chunks])
```

# Tratamos de obtener los verbos.

```python
# Ubica verbos
print("Verbos:", [token.lemma_ for token in doc if token.pos_ == "VERB"])
```

# Tratamos de obtener personas, organizacion u órganos, lugares y conceptos.

```python
# Ubica nombres, entidades o conceptos
for entity in doc.ents:
    print(entity.text, entity.label_)
```

# Resaltamos lo anterior en el documento.

```python
#Resaltar conceptos
from spacy import displacy
displacy.render(doc,style="ent",jupyter=True)
```

# Relaciona de forma gráfico palabras en el documento.

```python
#Grafos
from spacy import displacy
doc2 = list(doc.sents)
displacy.render(doc2,style="dep",jupyter=True,options={"distance":140})
```

# Extraemos organizaciones u órganos.

```python
# Ubica organizaciones u organos
for entity in doc.ents:
    if entity.label_ is "ORG":
     print(entity.text, entity.label_)
```

# Extraemos personas del documento.

```python
# Ubica nombres, entidades o conceptos
personas=[]
for entity in doc.ents:
    if entity.label_ == "PER":
      personas[]
     print(entity.text, entity.label_)
```
